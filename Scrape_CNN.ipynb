```python
import scrape_cnn as scrape
import pickle
```

```python
OUT_FILE = "output_data/output_cnn.pkl"
```

```python
results = {}

links = []
with open('sources/cnn_sources.txt', 'r') as sources:
    for line in sources:
        url = line.rstrip()
        links.append(url)
```

```python
links
```

```python
try:
    with open(OUT_FILE, 'rb') as handle:
        avg_sentiments = pickle.load(handle)
except FileNotFoundError:
    avg_sentiments = {}
```

```python
from datetime import datetime
del avg_sentiments[datetime(2011, 10, 7, 0, 0)]
```

```python
for i, url in enumerate(links):
    print(url)
    date = scrape.date_from_url(url)
    if date in avg_sentiments:
        print("Already have article from {}. Skipping {}".format(str(date)[:10], i))
        continue
    sentiments = scrape.analyze_sentiment_by_sentences(url, verbose=False)
    score = scrape.get_mean_sentiment(sentiments)
    headline = scrape.get_headline(url)
    print(headline)
    print("Finished source {} out of {}. Average score: {}".format(i + 1, len(links), score))
    avg_sentiments[date] = score
```

```python
sentiments
```

```python
from statistics import stdev, mean
mean_article = mean([avg_sentiments[x] for x in avg_sentiments])
std_article = stdev([avg_sentiments[x] for x in avg_sentiments])
print(mean_article, std_article)
```

```python
x = []
y = []
for item in avg_sentiments:
    x.append(str(item)[:10])
    y.append(avg_sentiments[item])
details = zip(x, y)
for x, y in sorted(details, key=lambda x: x[0]):
    if mean_article - 2 * std_article <= y <= mean_article + 2 * std_article:
        print("{},{}".format(x, y))
```

```python
with open(OUT_FILE, 'wb') as f:
    pickle.dump(avg_sentiments, f)
```
