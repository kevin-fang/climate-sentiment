```python
import scrape
```

```python
results = {}

links = []
with open('climate_change_sources.txt', 'r') as sources:
    for line in sources:
        url = line.rstrip()
        links.append(url)
```

```python
for i, url in enumerate(links):
    sentiments = scrape.analyze_sentiment_by_sentences(url, verbose=False)
    score = scrape.get_mean_sentiment(sentiments)
    date = scrape.date_from_url(url)
    print("Average score: {}".format(score))
    print("Finished source {} out of {}".format(i + 1, len(links)))
    results[date] = score
```
