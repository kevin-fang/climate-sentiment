```python
import scrape_fox as scrape
import requests
from lxml import html
import pickle
```

```python
OUT_FILE = "output_data/output_fox.pkl"
```

```python
links = set()
with open('sources/fox_sources.txt') as f:
    for url in f:
        links.add(url.strip().replace('.amp', ''))
```

```python
links
```

```python
def get_headline(url):
    page = requests.get(url)
    tree = html.fromstring(page.content)
    print("Finished downloading. Parsing...")

    divs = []
    for div in tree.xpath('//h1/text()'):
        divs.append(div)
    return divs[0]
```

```python
get_headline("https://www.foxnews.com/opinion/climate-summit-why-we-believe-paris-proposals-doom-billions-to-live-in-extreme-poverty")
```

```python
date_to_link = {}
for url in links:
    print(url)
    try:
        date = scrape.get_date(url)
        headline = get_headline(url)
        date_to_link[date] = headline
        
        print(date, url, headline)
    except (AttributeError, IndexError, ValueError):
        print("Missing date: {}".format(url))
        continue
```

```python
date_to_link
```

```python
OUT_FILE = "output_data/output_fox.pkl"
try:
    with open(OUT_FILE, 'rb') as handle:
        avg_sentiments = pickle.load(handle)
except FileNotFoundError:
    avg_sentiments = {}
    
for item in avg_sentiments:
    avg_sentiments[item] = (avg_sentiments[item], date_to_link[item])
```

```python
avg_sentiments
```

```python
with open("output_data/output_fox_with_headlines.pkl", 'wb') as f:
    pickle.dump(avg_sentiments, f)
```

```python
with open("output_data/output_fox_with_headlines.pkl", 'rb') as f:
    vals = pickle.load(f)
dates = []
headlines = []
sentiments = []

for date in vals:
    dates.append(str(date)[:10])
    headlines.append(vals[date][1].replace(",", ""))
    sentiments.append(vals[date][0])
```

```python
details = zip(dates, headlines, sentiments)
for x, y, z in sorted(details, key=lambda x: x[0]):
    print("{},{},{}".format(x, y, z))
```
