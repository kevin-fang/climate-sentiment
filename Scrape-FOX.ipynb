```python
import scrape_fox as scrape
import pickle
```

```python
OUT_FILE = "output_data/output_fox.pkl"
```

```python
results = {}

links = set()
with open('sources/fox_sources.txt', 'r') as sources:
    for line in sources:
        url = line.rstrip()
        links.add(url)
```

```python
links
```

```python
try:
    with open(OUT_FILE, 'rb') as handle:
        avg_sentiments = pickle.load(handle)
        print("Loaded from file")
except FileNotFoundError:
    avg_sentiments = {}
    print("Existing output not found. Creating new output file...")
```

```python
avg_sentiments
```

```python
for i, url in enumerate(links):
    print(url)

    try:
        date = scrape.get_date(url)
    except AttributeError:
        print("Missing date: {}".format(url))
        continue
    except (ValueError, IndexError):
        print("Link is incorrect or dead: {}".format(url))
        continue
    if date in avg_sentiments:
        print("Already have article from {}. Skipping {}".format(str(date)[:10], i))
        continue
    sentiments, date = scrape.analyze_sentiment_by_sentences(url, verbose=False)
    score = scrape.get_mean_sentiment(sentiments)
    print("Finished source {} out of {}. Average score: {}".format(i + 1, len(links), score))
    avg_sentiments[date] = score
```

```python
dead_links = []
for item in links:
    try:
        date = scrape.get_date(item)
    except (ValueError, IndexError):
        dead_links.append(item)
print(dead_links)
print(len(dead_links))
```

```python
with open(OUT_FILE, 'wb') as f:
    pickle.dump(avg_sentiments, f)
```

```python
x = []
y = []
for item in avg_sentiments:
    x.append(str(item)[:10])
    y.append(avg_sentiments[item])
details = zip(x, y)
for x, y in sorted(details, key=lambda x: x[0]):
    print("{},{}".format(x, y))
```

```python
from statistics import stdev, mean
mean_article = mean([avg_sentiments[x] for x in avg_sentiments])
std_article = stdev([avg_sentiments[x] for x in avg_sentiments])
print(mean_article, std_article)
```

```python
x = []
y = []
for item in avg_sentiments:
    x.append(str(item)[:10])
    y.append(avg_sentiments[item])
details = zip(x, y)
for x, y in sorted(details, key=lambda x: x[0]):
    if mean_article - 2 * std_article <= y <= mean_article + 2 * std_article:
        print("{},{}".format(x, y))
```
