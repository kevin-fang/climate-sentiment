```python
import scrape_economist as scrape
import pickle
import requests
from lxml import html
from datetime import datetime
```

```python
OUT_FILE = "output_data/output_economist.pkl"
```

```python
links = set()
with open('sources/economist_sources.txt') as f:
    for url in f:
        links.add(url.strip().replace('.amp', ''))
```

```python
links
```

```python
def get_headline(url):
    #base_url = "https://www.textise.net/showText.aspx?strURL="
    page = requests.get(url)
    tree = html.fromstring(page.content)
    print("Finished downloading. Parsing...")
    
    divs = []
    for div in tree.xpath('//title/text()'):
        divs.append(div.strip())
        
    return divs[0].split(' - ')[0]
```

```python
get_headline("https://www.economist.com/democracy-in-america/2016/01/25/the-republicans-and-climate-change")
```

```python
date_to_link = {}
for link in links:
    headline = get_headline(link).strip()
    date_to_link[scrape.date_from_url(link)] = headline
    print(link, headline)
```

```python
OUT_FILE = "output_data/output_economist.pkl"
try:
    with open(OUT_FILE, 'rb') as handle:
        avg_sentiments = pickle.load(handle)
except FileNotFoundError:
    avg_sentiments = {}
```

```python
for item in avg_sentiments:
    avg_sentiments[item] = (avg_sentiments[item], date_to_link[item])
```

```python
avg_sentiments
```

```python
with open("output_data/output_economist_with_headlines.pkl", 'wb') as f:
    pickle.dump(avg_sentiments, f)
```

```python
with open("output_data/output_economist_with_headlines.pkl", 'rb') as f:
    vals = pickle.load(f)
dates = []
headlines = []
sentiments = []

for date in vals:
    dates.append(str(date)[:10])
    headlines.append(vals[date][1].replace(",", ""))
    sentiments.append(vals[date][0])
```

```python
details = zip(dates, headlines, sentiments)
for x, y, z in sorted(details, key=lambda x: x[0]):
    print("{},{},{}".format(x, y, z))
```
