```python
import scrape_reuters as scrape
import pickle
import requests
from lxml import html
from datetime import datetime
```

```python
OUT_FILE = "output_data/output_reuters.pkl"
```

```python
results = {}

links = set()
with open('sources/reuters_sources.txt', 'r') as sources:
    for line in sources:
        url = line.rstrip()
        links.add(url)
```

```python
try:
    with open(OUT_FILE, 'rb') as handle:
        avg_sentiments = pickle.load(handle)
        print("Loaded from file")
except FileNotFoundError:
    avg_sentiments = {}
    print("Existing output not found. Creating new output file...")
```

```python
def get_headline(url):
    divs = scrape.get_url(url)
    delete = [" |"]
    divs = divs[8:]
    divs = filter(lambda x: x != '', divs)
    divs = filter(lambda x: "Min Read" not in x, divs)
    divs = filter(lambda x: x not in delete, divs)
    date_re = r'(\w+ \d+, \d+)'
    divs = list(divs)
    divs = divs[:-7]
    divs = divs[1:]
    return divs[0]
```

```python
date_to_link = {}
for link in links:
    print(link)
    
    try:
        date = scrape.get_date(link)
        headline = get_headline(link).strip()
        date_to_link[date] = headline
        print("Link: {}, date: {}, headline: {}".format(link, date, headline))
    except (AttributeError):
        print("Missing date: {}".format(url))
        continue
        
```

```python
not_in_date = []
for item in avg_sentiments:
    if item not in date_to_link:
        not_in_date.append(item)
        continue
    avg_sentiments[item] = (avg_sentiments[item], date_to_link[item])
for item in not_in_date:
    del avg_sentiments[date]
```

```python
with open("output_data/output_reuters_with_headlines.pkl", 'wb') as f:
    pickle.dump(avg_sentiments, f)
```

```python
with open("output_data/output_reuters_with_headlines.pkl", 'rb') as f:
    vals = pickle.load(f)
    
dates = []
headlines = []
sentiments = []

for date in vals:
    print(date)
    dates.append(str(date)[:10])
    headlines.append(vals[date][1].replace(",", ""))
    sentiments.append(vals[date][0])
```
